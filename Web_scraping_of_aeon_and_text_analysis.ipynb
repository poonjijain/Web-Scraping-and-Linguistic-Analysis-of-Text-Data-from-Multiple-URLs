```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re 
import nltk
nltk.download("all")

from nltk.stem import WordNetLemmatizer
```

```
from nltk.corpus import opinion_lexicon
pw= opinion_lexicon.positive()
```

```
def pos(text,pw):
  list1_lower = [str(x).lower() for x in text]
  list2_lower = [str(x).lower() for x in pw]
  matching = [x for x in list1_lower if x in list2_lower]
  return (len(matching))
```

```
nw=opinion_lexicon.negative()
```

```
def neg(text,nw):
  list1_lower = [str(x).lower() for x in text]
  list2_lower = [str(x).lower() for x in nw]
  matching = [x for x in list1_lower if x in list2_lower]
  return (len(matching))
```

```
def polarity_subjectivity(pos,neg,sw):
  ps=pos-neg/((pos+neg)+0.000001)
  sub=pos-neg/((sw)+0.000001)
  return(ps,sub)
```

```
def count(text):
  sentence_token = []
  avg_sent_len=0
  word_token = []
  if len(text) > 0:
    word_token=nltk.word_tokenize(text)
    sentence_token=nltk.sent_tokenize(text)
    if len(sentence_token)>0:
      avg_sent_len=len(word_token)/len(sentence_token)
  return(avg_sent_len)
```

```
def substitute(text):
  after=[]
  for w in text:
    res=re.sub(r'[^\w\s]',"",w) # not equal to alphanumeric symbols
    if res !="":
      after.append(res)
    
  return after
  
```

```
from nltk.corpus import stopwords
sw = set(stopwords.words('english'))
```

```
def eliminate(stopwords,clean_data):
  eliminated=[]
  for i in clean_data:
    c=0
    for x in stopwords:
      if str(i).lower() == str(x).lower():
        c=c+1
        continue
    if c==0:
        eliminated.append(i)
  return eliminated
```

```
def stemming(k):
  lemmatizer = WordNetLemmatizer()
  lem_list = []
  for word in k:
    if word.endswith("ed"):
        lemma = lemmatizer.lemmatize(word, pos='v')  # Assuming it's a verb
    elif word.endswith("es"):
        lemma = lemmatizer.lemmatize(word, pos='n')  # Assuming it's a noun
    else:
        lemma = word 
    
    lem_list.append(lemma)
  return(lem_list)
```

```
def syllable(words):
  c=0
  syl=0
  vowel=["a","i","e","o","u"]
  pattern = r"\b(" + "|".join(vowel) + r")\b"
  for i in words:
    match=0
    for j in i:
      if j in vowel:
        syl+=1
        match+=1
    if match>2:
      c+=1
  return (c,syl)
  
```

```

def pp(text):
  words = ["I", "we", "my", "ours", "us", "i", "We", "My", "Ours", "Us", "WE", "OURS", "MY"]
  word_counts=0
  pattern = r"\b(" + "|".join(words) + r")\b"
  matches = re.findall(pattern, text)
  for word in words:
    word_counts =word_counts + matches.count(word) 
  return(word_counts)
```

```
import requests 
import pandas as pd
from bs4 import BeautifulSoup
import nltk
 
url="https://aeon.co/psychology/social-psychology"
urls=[]
data=requests.get(url).text
soup=BeautifulSoup(data,"lxml")
dataframe=[]

all_urls=soup.find("div", class_="ThreeUp-sc-1ttdvtz-0 gweiIc")
for i in all_urls.find_all("a"):
    if i.get("href")[1]!="v":
        urls.append(i.get("href"))
for j in urls:
  l="https://aeon.co"+str(j)
  text = requests.get(l).text
  soup = BeautifulSoup(text, "lxml")
  content = soup.find("div", class_="styled__Highlightable-sc-1jhxu2u-3 gddtnP has-dropcap")
  paragraphs = soup.select("div.styled__Highlightable-sc-1jhxu2u-3.gddtnP.has-dropcap p")

# Store paragraph text in a list
  para = []
  t=""
  for paragraph in paragraphs:
    para.append(paragraph.get_text())
  for i in para:
    t+=i

# Print the paragraphs
  w=nltk.word_tokenize(t)
  avg=count(t)
  personal=pp(t)
  k=substitute(w)
  word_len=avg_word_len(k)
  lemmatized_list=stemming(k)
  com_count,syllable_count= syllable(lemmatized_list)
  
  if len(k)>0:
    syllablle_per_word=syllable_count/len(k)
    percent= com_count/len(k)
  fog_index=0.4*(avg+percent)
  no_stopw=eliminate(sw,k)
  word_count=(len(no_stopw))
  neg_count=neg(no_stopw,nw)
  pos_count=pos(no_stopw,pw)
  polarity_score,subjectivity_score=polarity_subjectivity(int(pos_count),int(neg_count),len(no_stopw))
  d={}
  d["URL"]=(str(url))
  d["POSITIVE SCORE"]=(pos_count)
  d["NEGATIVE SCORE"]=(neg_count)
  d["POLARITY SCORE"]=(polarity_score)
  d["SUBJECTIVITY SCORE"]=(subjectivity_score)
  d["AVG SENTENCE LENGTH"]=(avg)
  d["PERCENTAGE OF COMPLEX WORDS"]=(percent)
  d["FOG INDEX"]=(fog_index)
  d["AVG NUMBER OF WORDS PER SENTENCE"]=(avg)
  d["COMPLEX WORD COUNT"]=(com_count)
  d["WORD COUNT"]=(word_count)
  d["SYLLABLE PER WORD"]=(syllablle_per_word)
  d["PERSONAL PRONOUNS"]=(personal)
  d["AVG WORD LENGTH"]=(word_len)
  dataframe.append(d)
df=pd.DataFrame(dataframe)
print(df)
    
```
